# Método dos mínimos quadrados

## Recapitulando a Prática 2

Na Prática 2, aprendemos que gráficos são uma ferramenta bem interessante para explorar como duas grandezas se relacionam. Ao modelarmos qualquer fenômeno, gostaríamos de saber como uma grandeza $y$ se comporta quando variamos um parâmetro de controle $x$.

Para recordar, na prática 2, estudamos a deformação $x$ de uma barra metálica fixa em uma extremidade causada por uma força $F$ aplicada na extremidade oposta. Vimos que no limite elástico, vale a relação 
$$F = k x,$$

onde $k$ é a constante de proporcionalidade que depende dos parâmetros geométricos da barra (comprimento $L$, largura $b$, e espessura $d$) e de seu módulo de elasticidade $E$, uma propriedade intrínseca do material que a constitui. Explicitamente

$$F = \left( E \frac{d^3 b}{4L^3} \right) x$$.

No experimento, tínhamos controle da força peso $F$ aplicada (mudando a associação de massas) e também do comprimento $L$ da barra (mudando o ponto de fixação), então, analizamos dois casos

1. $x = k^{-1} F$ com $k = E \frac{d^3 b}{4L^3}$: que é uma relação linear. Ao plotarmos $F$ vs $x$ o coeficiente angular nos dá $k$.
2. $x = \kappa L^3$ com $\kappa = \frac{4 F}{E d^3 b}$: que é uma relação cúbica. Ao plotarmos $x = \kappa \ell$ com $\ell = L^3$ a relação $x$ vs $\ell$ volta a ser linear, com $\kappa$ sendo o coeficiente angular.


Para esses dois casos, nós traçamos manualmente retas que melhor representariam nosso conjunto de pontos e obtivemos os coeficientes $k$ e $\kappa$ escolhendo dois pontos:

1. $F = k x $ : $P_A = (x_1; F_1)$ e $P_B = (x_2; F_2)$ e tomamos $k = \frac{F_2-F_1}{x_2-x_1}$. 
2. $x = \kappa L^3 = \kappa \ell $: $P_A = (\ell_1; x_1)$ e $P_B = (\ell_2; x_2)$ e tomamos $\kappa =\frac{x_2-x_1}{\ell_2-\ell_1} $.


## Ajuste da melhor curva

Agora, nós vamos aprender um procedimento sistemático para encontrar a melhor curva que se ajusta a nossos dados experimentais. 

Vamos assumir que no nosso experimento queremos saber como uma grandeza $y$ se comporta quando variamos $x$ e que temos um modelo $y_{\textrm{model}} = f(x;\alpha_0, \alpha_1, ..., \alpha_M)$ que depende de alguns parâmetros $\alpha_0, \alpha_1, ..., \alpha_M$. A variável $x$ é conhecida como <i> variável independente </i>, equanto $y$ é a <i> variável dependente </i>. Feitas $n$ observações, temos um conjunto de $n$ pares $(x_i, y_i)$. Claro, o número de parâmetros depende do modelo. Podemos admitir que alguns dos $a_m$ vão ser nulos e tentar encontrar somente os coeficientes não-nulos. 

**Exemplos:**

Na parte 1. da Prática 2 anterior, nosso modelo era $F = k x$, observamos os pontos $(x_i, F_i)$ e nosso conjunto de parâmetros envolvia apenas $k$, o coeficiente angular.  


Poderíamos estar interessados em ajustar um <i>polinômio</i> de uma dada ordem em $x$ ao nosso conjunto de pontos. 

O modelo linear mais geral é
$$ y_{\textrm{linear}} = f(x;a,b) = b x + a ,$$
e tem $2$ parâmetros, $b$ que é o coeficiente angular e $a$ que é o coeficiente linear (isto é, é o ponto de intersecção da reta com o eixo $y$). *OBS*: na apostila, esses coeficientes estão trocados!

O modelo cúbico mais geral é
$$ y_{\textrm{cubic}} = f(x;a_0, a_1, a_2, a_3) = a_0 + a_1 x + a_2 x^2 + a_3 x^3, $$
e tem $4$ parâmetros, $a_0$ é o coeficiente linear, $a_1$ é o coeficiente associado à contribuição linear em $x$ para $y$, $a_1$ é o coeficiente associado à contribuição quadrática em $x$ para $y$, $a_3$ é o coeficiente associado à contribuição cúbica em $x$ para $y$.

Para um polinômio de ordem $P$, teríamos
$$ y_{\textrm{poly-P}} = f(x;a_0, a_1, a_2, ..., a_P) = \sum_{p=0}^P a_p x^p, $$
com $P+1$ parâmetros.


O **método dos mínimos quadrados**, como o próprio nome diz, se objetiva a encontrar o conjunto de parâmetros $\vec{a} = (a_0, a_1, ..., a_M)$ de um modelo $f(x; \vec{a})$ que minimiza o quadrado das distâncias $r_i = y_i - f(x_i; \vec{a})$ entre os pontos $y_i$ observados e o nosso modelo usando as observações $x_i$ como entrada. Para que isso valha para todos os pontos, nosso objetivo é minimizar o rezíduo

$$ S = \sum_{i=1}^{N}  r_i^2$$

para todos os parâmetros $\vec{\alpha}$. 

A condição para que $R^2$ seja mínimo é 

$$ \frac{\partial S}{\partial \alpha_m} = 0 \quad \quad \forall m,$$
onde $\partial \ \partial \alpha_m$ indica a derivada parcial tomada na variável $\alpha_m$.


Assim, teremos que calcular as derivadas parciais de $R$ para todos os parâmetros $\alpha_m$ do nosso modelo, igualar isto a $0$. Se tivermos $M+1$ parâmetros teremos $M+1$ equações para determiná-los. 

Parece abstrato? Vamos ver um exemplo: o ajuste linear $y_{\textrm{linear}} = f(x;a,b) = b x + a$! Temos dois parâmetros, assim duas derivadas parciais para calcular, uma é a $ \frac{\partial R^2}{\partial a}$ e a outra é a $ \frac{\partial R^2}{\partial b}$. Vamos assumir que temos $N$ pontos $(x_i,y_i)$. Assim, nosso rezíduo é

$$
\begin{align*}
    S & = \sum_{i=1}^N [y_i - f(x_i; a,b)]^2 \\
    & = \sum_{i=1}^N \left[ y_i - (b x_i + a) \right]^2.
\end{align*}
$$

Calculando as derivadas parciais

$$
\begin{align*}
    \frac{\partial S}{\partial b} & = -2 \sum_{i=1}^N \left[ y_i - (b x_i + a) \right] x_i 
    \\
    & = -2 \sum_{i=1}^N y_i x_i + 2 b \sum_{i=1}^N x_i^2 + 2 a \sum_{i=1}^N x_i
    \\
    & = -2 S_{xy} + 2 b S_{xx} + 2 a S_x
\end{align*}
$$

$$
\begin{align*}
    \frac{\partial S}{\partial a} & = -2 \sum_{i=1}^N \left[ y_i - (b x_i + a) \right] 
    \\
    & = -2 \sum_{i=1}^N y_i + 2 b \sum_{i=1}^N x_i + 2 \sum_{i=1}^N a
    \\
    & = -2 S_{y} + 2 b S_{x} + 2  a N
\end{align*}
$$


Então, temos duas equações para determinar $a$ e $b$:
$$\begin{align*}
     b S_{xx} + a S_x & = S_{xy}
    \\
     b S_{x} +  a N & = S_{y}
\end{align*}
$$

Resolvendo para $a$
$$\begin{align*}
    & & b S_{xx} + a S_x & = S_{xy} & \times N
    \\
    +& &\\
    & & b S_{x} +  a N & = S_{y} & \times -S_{x}
    \\
    = & & & &
    \hline
    \\
    & & b S_{xx} N + a S_x N & = S_{xy} N &
    \\
    +& &\\
    & & -b S_{x}^2 +  - a S_x N & = -S_{y}S_x & 
        \\
    = & & & &
    \hline
    \\
    & & b (S_{xx} N - S_x^2 ) & = S_{xy} N - S_y S_x
\end{align*}
$$
o que resulta em 
$$
    b = \frac{S_{xy} N - S_y S_x}{S_{xx} N - S_x^2 }.
$$

Resolvendo para $b$:

$$\begin{align*}
    & & b S_{xx} + a S_x & = S_{xy} & \times S_x
    \\
    +& &\\
    & & b S_{x} +  a N & = S_{y} & \times -S_{xx}
    \\
    = & & & &
    \hline
    \\
        & & b S_{xx} S_x + a S_x S_x & = S_{xy} S_x &
    \\
    & & -b S_{x} S_{xx} +  - a N S_{xx} & = -S_{y} S_{xx} &     
    \\
    = & & & &
    \hline
    \\
        & & a (S_x^2 - NS_{xx}) & = S_{xy} S_x - S_{xx}S_{y} &
\end{align*}
$$
o que resulta em

$$
a = \frac{S_{xx}S_{y}-S_{xy} S_x}{NS_{xx}-S_x^2}.
$$

Mas, veja que tínhamos já a equação $a S_{x} +  b N  = S_{y}$ para $b$ em função de $a$! Lembrando que as médias dos valores $x$ e $y$ são 

$$
\begin{align}
    S_x / N & = \bar{x} \\
    S_y / N & = \bar{y} \\
\end{align}
$$

temos 
$$
a = \bar{y} - b \bar{x},
$$
bem mais simples se já tivermos calculado o $b$!

**A qualidade do nosso modelo** é dada pelo chamado coeficiente de determinação $R^2$. 
$$
R^2 = 1 - \frac{\sum_j r_j^2}{\sum_j (y_i-\bar{y})^2}.
$$


Se substituirmos os valores de $a$ e $b$ encontrados acima, temos
$$
R^2  = \frac{(N S_{xy} - S_x S_y)^2}{(N S_{xx} - S_x^2)(N S_{yy} - S_y^2)}
$$
ou, ainda 
$$
R = \frac{\sum_{i=1}^N (x_i-\bar{x}) (y_i - \bar{y})}{\sqrt{\sum_{j=1}^N (x_j-\bar{x})^2} \sqrt{\sum_{l=1}^N (y_l-\bar{y})}},
$$
que é conhecido como coeficiente de correlação ou coeficiente de Pearson.

Quanto mais próximo da unidade for o $R \rightarrow 1$, mais fideldigno é o nosso modelo ao conjunto de pontos observados. 


**E os erros nesses parâmetros do ajuste **?

Bem, a distância do ponto observado $y_i$ para $x_i$ e o valor que ele teria de acordo com o nosso modelo $y_i' = b x_i + a$ é 
$$
r_i = y_i - y_i'
$$

Existe um intervalo de confiança para $a$ e $b$ que assume que os erros do ajuste são distribuídos de acordo com uma probabilidade Gaussiana. 
estimador para a variância deste ponto é
$$
\Delta y = \sqrt{\sum_{i=1}^N \frac{r_i^2}{N-2}}.
$$

Os erros para os coeficientes $a$ e $b$ serão
$$
\Delta b = \frac{\Delta y}{\sqrt{ S_{xx} - \frac{S_x^2}{N}}}
$$
e
$$
\Delta a = \Delta b \sqrt{\frac{S_{xx}}{N}}
$$

### Em um futuro breve
O que é bacana (hold on, vocês verão isso em álgebra linear) é que dá para chegar neste mesmo conjunto de equações de forma matricial. Com um sisteminha
$$
\begin{align*}
    \begin{pmatrix} 
    S_{xy} \\
    S_y 
    \end{pmatrix}
    & = 
    \begin{pmatrix} 
    S_{xx} & S_x \\
    S_x  & N 
    \end{pmatrix}
        \begin{pmatrix} 
    b \\
    a 
    \end{pmatrix}
\end{align*}
$$

chamando a matriz 
$$
\begin{align*}
    \mathcal{S} & = 
    \begin{pmatrix} 
    S_{xx} & S_x \\
    S_x  & N 
    \end{pmatrix}
\end{align*}
$$
e admitindo que existe a inversa $\mathcal{S}^{-1}$, temos 
$$
\begin{align*}
    \begin{pmatrix} 
    S_{xy} \\
    S_y 
    \end{pmatrix}
    & = \mathcal{S}
        \begin{pmatrix} 
    b \\
    a 
    \end{pmatrix}
    \\
        \begin{pmatrix} 
    b \\
    a 
    \end{pmatrix}  & = 
    \mathcal{S}^{-1} 
        \begin{pmatrix} 
    S_{xy} \\
    S_y 
    \end{pmatrix}
\end{align*}
$$

que elegante!

Outra forma ainda mais elegante que usa uma solução matricial para este problema é a seguinte: consideremos os conjuntos $y_i$ formando um vetor e os coeficientes $a$ e $b$ também formando um vetor. Para todos os nossos $N$ pontos observados no experimento nosso modelo segue

$$
\begin{align*}
    \begin{pmatrix} 
    y_1 \\
    y_2 \\
    \vdots
    \\
    y_N
    \end{pmatrix}
    & = 
    \begin{pmatrix} 
    1 & x_1  \\
    1 & x_2 \\
    1 & \vdots \\
    1 & x_N
    \end{pmatrix}
    \begin{pmatrix} 
    b \\
    a 
    \end{pmatrix} 
\end{align*}
$$

o que pode ser sumarizado na equação

$$
    \vec{y}  = X \vec{p},
$$

com $X$ sendo a matriz de dimensão $N \times 2$ e $\vec{p} = (b a)$ sendo um vetor coluna de dimensão $2 \times 1$, e $\vec{y}$ um vetor coluna de dimensão $N \times 1$. Se $X$ tem uma transporta $X^T$ e a combinação $X^T X$ tem uma inversa $(X^T X)^{-1}$, segue que 

$$
   \vec{p}  = (X^T X)^{-1} X^T \vec{y}
$$

#### Ajuste linear brute-force
```python
# importanto a biblioteca numpy
import numpy as np

#    Ajuste linear com numpy
def ajuste_linear_bf(x_i:np.array, y_i:np.array, N:int):
    '''
        ajuste_linear_bf

        Inputs:
            x_i: valores dos dados x (np.array)
            y_i: valores dos dados y (np.array)
            N: número de pontos
    '''
    # somas
    Sx = x_i.sum()
    Sy = y_i.sum()
    Sxx = (x_i * x_i).sum()
    Sxy = (x_i * y_i).sum()
    Syy = (y_i * y_i).sum()

    # valores médios
    y_m = Sy / N
    x_m = Sx / N

    # cálculo dos coeficientes
    b = (Sxy * N - Sy * Sx) / (Sxx * N - Sx**2)
    a = y_m - b * x_m

    # valores y_Ci para o modelo
    y_Ci = a + b * x_i

    # distância quadrada dos pontos y_i com os do modelo y_Ci
    r_i = y_i - y_Ci

    # estimador da variância
    Delta_y = np.sqrt((r_i**2).sum() / (N-2))

    # erro nos coeficientes
    Delta_b = Delta_y / np.sqrt(Sxx-Sx**2 / N)
    Delta_a = Delta_b * np.sqrt(Sxx / N)


    # R quadrado
    R2 = 1 - Delta_y**2 * (N-2) / ((y_i - y_m)**2).sum()

    return [a,b], [Delta_a, Delta_b], R2


```

                
#### Ajuste linear com bibliotecas
```python
# importando as funções
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# modelo linear
def modelo_linear(x, b:float, a:float):
    '''
        modelo_linear

        Inputs:
        x: variável independente
        a: coeficiente linear
        b: coeficiente angular
    '''
    return b*x + a

#   Ajuste linear com python libraries scipy e R2 com sklearn
def ajuste_linear_pl(x_i:np.array, y_i:np.array, N:int):
    '''
        ajuste_linear_pl

        Inputs:
            x_i: valores dos dados x (np.array)
            y_i: valores dos dados y (np.array)
            N: número de pontos
    '''

    coefs_lin, cov_lin = curve_fit(modelo_linear, x_i, y_i)
    coefs_err = np.sqrt(np.diag(cov_lin))

    # cálculo dos coeficientes: aqui a ordem é a mesma em que eles aparecem na função
    b = coefs_lin[0]
    a = coefs_lin[1]

    # valores y_Ci para o modelo
    y_Ci = a + b * x_i

    # erro nos coeficientes
    Delta_b = coefs_err[0]
    Delta_a = coefs_err[1]


    # R quadrado
    R2 = r2_score(y_Ci, y_i)

    return [a,b], [Delta_a, Delta_b], R2

```

# Prática 3

Na prática 3 nós vamos medir a aceleração da gravidade $g$ de duas formas diferentes:

1. **Oscilação de um pêndulo**: no limite de pequenos ângulos, o período do pêndulo depende somente do comprimento e da aceleração da gravidade

$$
    T = 2\pi \sqrt{\frac{L}{g}}
$$

Iremos variar o comprimento do pêndulo $L$ e medir o período médio referente a $N=10$ oscilações $T = t_{N} / N$. Graficando $T^2$ vs $L$, isto é,

$$
    T^2 = 4\pi^2 \frac{L}{g}
$$

o coeficiente angular 

$$b = \frac{4\pi^2}{g}$$ 

nos permitirá obter a aceleração da gravidade.

Note que o erro na medida de $T$ é o tempo de reação que contabilizamos como sendo o intervalo entre o observador ligar ou desligar o cronômetro. Já o erro de $L$ é o erro da trena, que mede $1$ mm. Esperamos que qualquer coeficiente linear $a$ resultante do ajuste $y=bx +a$ seja próximo de zero.

Veja que nossa constante $b$ é esperada ser próxima de $4$, já que $\pi^2 \sim 9.87$!


2. **Plano inclinado**: no eixo paralelo ao plano, o movimento é unidirectional. Nessa direção, o movimento é ditado pela equação horária

$$
    y(t) = v_0 t + a \frac{t^2}{2}
$$

Iremos coletar as posições $y(t_i)$ ao longo de uma fita que é termicamente marcada por um faiscador a uma frequência  $f = 5$ Hz. O intervalo de tempo é fixo em $t=f^{-1} = 0,2$s e, portanto, $t_i = i \times 0,2 s$.

Aqui, vamos considerar o erro nas medidas de $y$ como sendo o erro da trena $1$mm. O erro do faiscador será desconsiderado. 

Para fazer o ajuste linear $y = \alpha + \beta x$, graficaremos 

$$
\frac{y}{t} = v_0 + \frac{a}{2} t
$$

e identificaremos os coeficientes angular $\beta = \frac{g\sin{\theta}}{2}$ e linear $v_0$.

Determinaremos a aceleração da gravidade a partir de $\beta$.

````python
# força os gráficos a aparecerem inline
%matplotlib inline

# pacotes numpy e pandas
import numpy as np
import pandas as pd

# lendo os dados em uma tabela via pandas
dados1 = pd.read_csv('dados_pratica3_tabela1.csv')
N1 = dados1.shape[0]
dados1
````

````python
# construindo uma tabela auxiliar seguindo o modelo da tabela 3.1 da página 52 da apostila

T_i = dados1['$$t_N(\pm0.2s)$$']/10
L_i = dados1['$$L(\pm0.001m)$$']

x_i = L_i
y_i = T_i**2

````

````python
x_m = x_i.mean()
y_m = y_i.mean()

x_i2 = x_i**2
x_i_m_xm = (x_i -x_m)
xi_m_xm_yi = x_i_m_xm * y_i

tabela1 = pd.DataFrame()
# Força em N
tabela1[r'$$i$$'] = np.arange(1,N1+1,1)
tabela1[r'$$x_i$$'] = x_i
tabela1[r'$$y_i$$'] = y_i
tabela1[r'$$x_i -\bar{x}$$'] = x_i_m_xm
tabela1[r'$$x_i^2$$'] = x_i**2
tabela1[r'$$(x_i -\bar{x})y_i$$'] = xi_m_xm_yi
tabela1[r'$$(x_i -\bar{x})^2$$'] = x_i_m_xm**2

tabela1
````

````python
# importando as funções do plot linear
from funcoes_ajuste_linear import ajuste_linear_bf, ajuste_linear_pl
````

````python
# coeficientes brute force
[a,b], [Delta_a, Delta_b], R2 = ajuste_linear_bf(x_i, y_i, x_i.shape[0])


print(f' b = {b}+-{Delta_b} \n a = {a}+-{Delta_a}\n R2 = {R2}')
````

````python
# coeficientes funções python
[a,b], [Delta_a, Delta_b], R2 = ajuste_linear_pl(x_i, y_i, x_i.shape[0])


print(f' b = {b}+-{Delta_b} \n a = {a}+-{Delta_a}\n R2 = {R2}')
````

````python
# modelo
y_Ci = a + b * x_i

tabela1[r'$$y_{Ci}$$'] = y_Ci

tabela1[r'$$(y_{Ci}-y_i)^2$$'] = (y_Ci - y_i)**2

tabela1.loc[r'$$\sum_{i=1}^N$$'] = tabela1.sum()
tabela1
````

````python
# plotando os gráficos 
from graficos import *

````

````python
fig1, ax1 = grafico_papel_milimetrado(x_i, y_i)

xfit = np.linspace(-1, math.ceil(x_i.max())+1, 101)

ax1.plot(xfit, a + b * xfit, 'r-', label='fit')

ax1.set_xlabel(r'$L(\pm 10^{-3}m)$')
ax1.set_ylabel(r'$T^2(\pm 10^{-1}s^2)$')

leg1 = ax1.legend()

plt.show()
````
